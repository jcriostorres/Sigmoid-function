````python
import numpy as np
import matplotlib.pyplot as plt
from __future__ import division

#Sigmoidal
def sigmoidal(z):
    
    #Complete la siguiente línea con el código para calcular la salida de la función sigmoidal
    #s = 
    return 1/ (1+np.exp(-z))
    
    #return s
    

#Escriba el código para realizar la gráfica de la función aquí

z = np.linspace(-5,5,50)
plt.plot(z, sigmoidal(z))
plt.title("Sigmoidal activation Function")
plt.xlabel("z")
plt.ylabel("y")
plt.grid(True)
plt.show()
````

# Cuándo utilizar la regresión logística #
La regresión logística funciona muy bien para ***clasificadores binarios***. Para aclarar, la clasificación binaria se refiere al proceso de predecir uno de dos resultados posibles. Por ejemplo, si intentamos predecir si un correo electrónico es spam o no, se trata de un problema de clasificación binaria porque ***sólo hay dos resultados posibles: "spam" o "no spam".***
Sin embargo, existen ciertas **limitaciones a la regresión logística**. En particular, puede tener __dificultades cuando lidiar con grandes conjuntos de datos de calidad inferior__. Por ejemplo, considere un conjunto de datos lleno de numerosos valores faltantes, valores atípicos o características irrelevantes. En estas circunstancias, el modelo de regresión logística podría tener dificultades para producir predicciones precisas.
Además, si bien la regresión logística puede manejar relaciones lineales entre características y la variable objetivo de manera efectiva, puede quedarse corta cuando se trata de relaciones complejas y no lineales. Imagine un conjunto de datos donde la relación entre las variables predictivas y el objetivo no es una línea recta sino una curva; un modelo de regresión logística podría tener dificultades en tales escenarios.
A pesar de estas limitaciones,  _**la regresión logística a menudo puede servir como un punto de partida sólido para tareas de clasificación. Proporciona un rendimiento de referencia que se puede utilizar para comparar la eficacia de modelos más complejos. Incluso si no ofrece la mayor precisión, sí ofrece interpretabilidad y simplicidad, lo que puede resultar valioso en determinados contextos.**_
